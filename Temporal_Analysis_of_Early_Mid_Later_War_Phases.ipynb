# **2. Data Preprocessing**

### **2.1 Combine Datasets**

```python
df_all = pd.concat([df_early, df_mid, df_later])
print("\nTotal Final Tweets:", len(df_all))

df_all
```

### **2.2 Language Distribution**

```python
lang_counts = (df_all.groupby(['language', 'language_full']).size().reset_index(name='count').sort_values(by='count', ascending=False))
lang_counts['proportion(%)'] = (lang_counts['count'] / lang_counts['count'].sum() * 100).round(2)

lang_counts
```

# **3. Sentiment Analysis**

### **3.1 Sentiment Summary**

```python
phase_order = ['Early War', 'Mid War', 'Later War']

sentiment_counts = pd.crosstab(df_all['phase'], df_all['sentiment']).loc[phase_order]
sentiment_percent = (pd.crosstab(df_all['phase'], df_all['sentiment'], normalize='index').loc[phase_order] * 100).round(2)
combined_sentiment_table = pd.concat([sentiment_counts, sentiment_percent.round(2)], axis=1, keys=['SentimentCount', 'Proportion(%)'])

combined_sentiment_table
```

### **3.2 Datewise Trend by Sentiment**

```python
daily_sentiment_all = (df_all.groupby(['phase', 'date', 'sentiment']).size().reset_index(name='count'))
daily_sentiment_all['phase'] = pd.Categorical(daily_sentiment_all['phase'], categories=phase_order, ordered=True)
daily_sentiment_all = daily_sentiment_all.sort_values(['phase', 'date'])
daily_sentiment_all['proportion(%)'] = (daily_sentiment_all.groupby(['phase', 'date'])['count'].transform(lambda x: (x / x.sum()) * 100).round(2))

daily_sentiment_all
```

### **3.3 Retweets Summary by Sentiment**

```python
retweets_sentiment_count = (df_all.groupby(['phase', 'sentiment'])['retweetcount'].sum().unstack('sentiment').loc[phase_order])
retweets_sentiment_percent= (retweets_sentiment_count.div(retweets_sentiment_count.sum(axis=1), axis=0) * 100).round(2)
combined_retweets_sentiment_table = pd.concat([retweets_sentiment_count, retweets_sentiment_percent.round(2)], axis=1, keys=['RetweetsCount', 'Proportion(%)'])

combined_retweets_sentiment_table
```

### **3.4 Datewise Retweets Trend by Sentiment**

```python
daily_retweet_sentiment = (df_all.groupby(['phase', 'date', 'sentiment'])['retweetcount'].sum().reset_index(name='retweetcount'))
daily_retweet_sentiment['phase'] = pd.Categorical(daily_retweet_sentiment['phase'], categories=phase_order, ordered=True)
daily_retweet_sentiment = daily_retweet_sentiment.sort_values(['phase', 'date'])
daily_retweet_sentiment['proportion(%)'] = (daily_retweet_sentiment.groupby(['phase', 'date'])['retweetcount'].transform(lambda x: (x / x.sum()) * 100).round(2))

daily_retweet_sentiment
```

### **3.5 Likes Summary by Sentiment**

```python
likes_sentiment_count = (df_all.groupby(['phase', 'sentiment'])['favorite_count'].sum().unstack('sentiment').loc[phase_order])
likes_sentiment_percent= (likes_sentiment_count.div(likes_sentiment_count.sum(axis=1), axis=0) * 100).round(2)
combined_likes_sentiment_table = pd.concat([likes_sentiment_count, likes_sentiment_percent.round(2)], axis=1, keys=['LikesCount', 'Proportion(%)'])

combined_likes_sentiment_table
```

### **3.6 Datewise Likes Trend by Sentiment**

```python
daily_like_sentiment = (df_all.groupby(['phase', 'date', 'sentiment'])['favorite_count'].sum().reset_index(name='favorite_count'))
daily_like_sentiment['phase'] = pd.Categorical(daily_like_sentiment['phase'], categories=phase_order, ordered=True)
daily_like_sentiment = daily_like_sentiment.sort_values(['phase', 'date'])
daily_like_sentiment['proportion(%)'] = (daily_like_sentiment.groupby(['phase', 'date'])['favorite_count'].transform(lambda x: (x / x.sum()) * 100).round(2))

daily_like_sentiment
```

# **7. Statistical and Effect Analysis**

### **7.1 Chi-Square Test**

```python
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(sentiment_counts)

print("Sentiment vs. Phase")
print(f"Chi-square: {chi2:.2f}")
print(f"p-value: {p:.3f}")
print(f"Degrees of Freedom: {dof}")

chi2_e, p_e, dof_e, expected_e = chi2_contingency(emotion_counts)

print("\nEmotion vs. Phase")
print(f"Chi-square: {chi2_e:.2f}")
print(f"p-value: {p_e:.3f}")
print(f"Degrees of Freedom: {dof_e}")
```

### **7.2 Cramer's V Size Effect**

```python
import numpy as np

cramers_v_sent = np.sqrt(chi2 / (sentiment_counts.sum().sum() * (min(sentiment_counts.shape) - 1)))
cramers_v_emot = np.sqrt(chi2_e / (emotion_counts.sum().sum() * (min(emotion_counts.shape) - 1)))

print("Sentiment vs. Phase")
print(f"Cramer's V: {cramers_v_sent:.3f}")
print("\nEmotion vs. Phase")
print(f"Cramer's V: {cramers_v_emot:.3f}")
```
