# **2. Data Preprocessing**

### **2.1 Explore Data**

```python
print("Column Names:\n\n", df.columns)
```

```python
columns = ['text', 'language', 'tweetcreatedts', 'retweetcount', 'favorite_count', 'hashtags', 'extractedts']

df_later = df[columns].copy()
df_later.head(10)
```

### **2.2 Date Range**

```python
print("Maximum Date:", df_later['extractedts'].min())
print("Minimum Date:", df_later['extractedts'].max())
```

### **2.3 Timestamp Feature Extraction**

```python
df_later.rename(columns={'extractedts': 'timestamp'}, inplace=True)
df_later['timestamp'] = pd.to_datetime(df_later['timestamp'])
df_later['date'] = df_later['timestamp'].dt.date
df_later['hour'] = df_later['timestamp'].dt.hour
df_later['phase'] = 'Later War'
df_later[['timestamp', 'date', 'hour', 'phase']].head(10)
```

### **2.4 Data Cleaning**

```python
import re

def clean_tweet(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+", "", text)
    text = re.sub(r"@\w+", "", text)
    text = re.sub(r"#", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df_later['clean_text'] = df_later['text'].apply(clean_tweet)
df_later.drop_duplicates(subset='clean_text', inplace=True)
df_later.dropna(subset=['clean_text'], inplace=True)
df_later = df_later[df_later['clean_text'].str.strip() != '']
df_later.reset_index(drop=True, inplace=True)

print("\nTotal Tweets:", len(df_later))
df_later[['clean_text']].head(10)
```

### **2.5 Language Mapping**

```python
lang_map = {
    'en': 'English', 'es': 'Spanish', 'und': 'Undefined', 'de': 'German', 'fr': 'French', 'it': 'Italian',
    'uk': 'Ukrainian', 'ru': 'Russian', 'in': 'Indonesian', 'nl': 'Dutch', 'pt': 'Portuguese', 'tr': 'Turkish',
    'hi': 'Hindi', 'ja': 'Japanese', 'th': 'Thai', 'ro': 'Romanian', 'ar': 'Arabic', 'sv': 'Swedish',
    'fi': 'Finnish', 'el': 'Greek', 'pl': 'Polish', 'tl': 'Tagalog', 'et': 'Estonian', 'ta': 'Tamil',
    'ca': 'Catalan', 'vi': 'Vietnamese', 'ur': 'Urdu', 'eu': 'Basque', 'ht': 'Haitian Creole', 'ko': 'Korean',
    'iw': 'Hebrew', 'da': 'Danish', 'te': 'Telugu', 'cs': 'Czech', 'zh': 'Chinese', 'fa': 'Persian',
    'lv': 'Latvian', 'bn': 'Bengali', 'kn': 'Kannada', 'sl': 'Slovenian', 'lt': 'Lithuanian', 'no': 'Norwegian',
    'sr': 'Serbian', 'gu': 'Gujarati', 'bg': 'Bulgarian', 'ka': 'Georgian', 'cy': 'Welsh', 'is': 'Icelandic',
    'mr': 'Marathi', 'hu': 'Hungarian', 'ml': 'Malayalam', 'km': 'Khmer', 'ne': 'Nepali', 'hy': 'Armenian', 'my': 'Burmese'}


df_later['language_full'] = df_later['language'].map(lang_map)
lang_counts = (df_later.groupby(['language', 'language_full']).size().reset_index(name='count').sort_values(by='count', ascending=False))
lang_counts['proportion(%)'] = (lang_counts['count'] / lang_counts['count'].sum() * 100).round(2)
lang_counts
```

# **3. Sentiment Analysis**

### **3.1 Sentiment Classification Pipeline**

```python
from transformers import pipeline

sentiment_model = pipeline(
    "sentiment-analysis",
    model="cardiffnlp/twitter-xlm-roberta-base-sentiment",
    device=0)
```

```python
texts = df_later['clean_text'].astype(str).str.slice(0,512).tolist()

def get_sentiment_batch(texts, batch_size=32):
    results = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        batch_results = sentiment_model(batch)
        results.extend(batch_results)

    return results

sentiment_results = get_sentiment_batch(texts, batch_size=32)

sentiments = []
sent_scores = []

for result in sentiment_results:
    sentiments.append(result['label'])
    sent_scores.append(result['score'])

df_later['sentiment'] = sentiments
df_later['sentiment_score'] = sent_scores
df_later[['clean_text','sentiment','sentiment_score']].head(10)
```

### **3.2 Sentiment Summary**

```python
sentiment_summary = (df_later['sentiment'].value_counts().reset_index())
sentiment_summary['proportion(%)'] = (sentiment_summary['count'] / sentiment_summary['count'].sum() * 100).round(2)

sentiment_summary
```

# **4. Emotion Analysis**

### **4.1 Emotion Classification Pipeline**

```python
emotion_model = pipeline(
    "zero-shot-classification",
    model="MoritzLaurer/mDeBERTa-v3-base-xnli-multilingual-nli-2mil7",
    device=0)
```

```python
candidate_labels = ["joy", "sadness", "anger", "fear", "love", "surprise"]

def get_emotions_batch(texts, batch_size=64):
    results = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]
        outputs = emotion_model(batch, candidate_labels)
        results.extend(outputs)

    return results

emotion_results = get_emotions_batch(texts, batch_size=64)

emotions = []
emotion_scores = []

for result in emotion_results:
    emotions.append(result['labels'][0])
    emotion_scores.append(result['scores'][0])

df_later['emotion'] = emotions
df_later['emotion_score'] = emotion_scores
df_later[['clean_text', 'emotion', 'emotion_score']].head(10)
```

### **4.2 Emotion Summary**

```python
emotion_summary = (df_later['emotion'].value_counts().reset_index())
emotion_summary.columns = ['emotion', 'count']
emotion_summary['proportion(%)'] = (emotion_summary['count'] / emotion_summary['count'].sum() * 100).round(2)

emotion_summary
```
